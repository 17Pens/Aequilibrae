```python
from copy import copy
import json
import geopandas as gpd
import shapely

# table = copy(links[['speed', 'geometry']])
# table2 = gpd.GeoDataFrame(columns = ['speed', 'geometry', 'at'])
new_list = []
for index, row in links.iterrows():
    print()
    print(index)
    a_node = row['a_node']
    b_node = row['b_node']

    if row['speed'] != None:
        speed = json.loads(row['speed'])
        # print(type(speed))
        if type(speed) == dict:
            # print(speed)
            new_list.append({'speed': speed['maxSpeed'][0],'a_node': a_node,'b_node': b_node})
            print('dict')
        elif type(speed) == list and len(speed) >= 1:
            # at_values = sorted(list(set([at for entry in speed for at in entry['at']])))

            # Extract the 'at' list from each dictionary
            # eg [[0.0, 0.064320774], [0.064320774, 1.0]]
            at_values_list = [entry['at'] for entry in speed]

            # Calculate differences between consecutive numbers in each 'at' list. This list iterates through each 'at' 
            # list in at_values_list and calculates the difference between consecutive elements using (at[i + 1] - at[i]).
            # The result is a flat list of differences for all 'at' lists.
            # eg [0.064320774, 0.935679226]
            differences = [diff for at in at_values_list for diff in (at[i + 1] - at[i] for i in range(len(at) - 1))]
             
            # Create a dictionary with keys as flat_differences and values as at_values_list
            result_dict = dict(zip(differences, at_values_list))
            print(result_dict)
       
            max_difference = max(differences)

            if max_difference >= 0.7:
                # Find the index of the value in the differences list
                index = differences.index(max_difference)

                # Access the corresponding entry in the original 'data' list to access the 'maxSpeed' value
                max_speed = speed[index]['maxSpeed'][0]

                new_list.append({'speed': max_speed, 'at': result_dict[max_difference], 'a_node': a_node,'b_node': b_node})
                print(result_dict[max_difference])
            else:
                print('splitting')
                for element in differences:
                    # Find the index of the value in the differences list
                    index_d = differences.index(element)
                    print(index_d)

                    # Access the corresponding entry in the original 'data' list to access the 'maxSpeed' value
                    max_speed = speed[index_d]['maxSpeed'][0]
                    print(max_speed)
                    new_list.append({'speed': max_speed, 'at': result_dict[element] ,'a_node': a_node,'b_node': b_node})
        
    else:
        new_list.append({'speed': row['speed'],'a_node': a_node,'b_node': b_node})
        print('none')


gdf = gpd.GeoDataFrame(new_list)
gdf['speed']
```

```python
from shapely.geometry import LineString, Point

def find_location(linestring, percentage):
    total_length = linestring.length
    target_distance = total_length * percentage
    return linestring.interpolate(target_distance)

def trim_geometry(node_lu, row, at_specifications):
    lat_long_a = node_lu[row["a_node"]]['coord']
    lat_long_b = node_lu[row["b_node"]]['coord']
    
    # Convert LineString to a list of coordinates
    coords = list(row.geometry.coords)

    split_line_strings = []

    for start_percentage, end_percentage in at_specifications:
        start_point = find_location(LineString(coords), start_percentage)
        end_point = find_location(LineString(coords), end_percentage)

        start, end = -1, -1

        for j, coord in enumerate(coords):
            if start_point.equals(Point(coord)):
                start = j
            if end_point.equals(Point(coord)):
                end = j

        if start < 0 or end < 0:
            raise RuntimeError("Couldn't find the start or end coords in the given linestring")

        split_line_strings.append(LineString(coords[start:end + 1]))

    return split_line_strings

# Example usage
at_specifications = [(0.0, 0.064320774), (0.064320774, 1.0)]

# Replace with your actual row and node_lu
row = {"a_node": 1, "b_node": 2, "geometry": LineString((153.18881 -27.6837219), (153.1888603 -27.6838145), (153.1889368 -27.6839668), (153.1891584 -27.6844079))}
node_lu = {1: {'coord': (0, 0)}, 2: {'coord': (3, 3)}}

result = trim_geometry(node_lu, row, at_specifications)
print(result)
```

```python
from shapely.geometry import LineString, Point

def split_linestring_at_percentage(linestring, percentage):
    total_length = linestring.length
    target_distance = total_length * percentage
    split_point = linestring.interpolate(target_distance)
    print(split_point)

    # Convert the CoordinateSequence to a list of tuples
    coords = list(linestring.coords)

    # If the split point is not in the original coordinates, add it
    if tuple(split_point.coords[0]) not in coords:
        coords.append(tuple(split_point.coords[0]))

    # Sort the coordinates based on distance
    coords.sort(key=lambda x: linestring.project(Point(x)))

    # Find the index of the split point in the sorted list
    split_index = coords.index(tuple(split_point.coords[0]))

    # Split the linestring into two parts
    part1 = LineString(coords[:split_index + 1])
    part2 = LineString(coords[split_index:])

    return part1, part2

# Example usage
linestring_coords = [
    (153.18881, -27.6837219),
    (153.1888603, -27.6838145),
    (153.1889368, -27.6839668),
    (153.1891584, -27.6844079)
]

linestring = LineString(linestring_coords)

percentage = 0.064320774
part1, part2 = split_linestring_at_percentage(linestring, percentage)

print(part1)
print(part2)
```

```python
# We grab all the links data as a Pandas DataFrame so we can process it easier
# links = project.network.links.data

# We create a Folium layer
network_links = folium.FeatureGroup("links")

# We do some Python magic to transform this dataset into the format required by Folium
# We are only getting link_id and link_type into the map, but we could get other pieces of info as well
for i, row in links.iterrows():
    points = row.geometry.wkt.replace("LINESTRING ", "").replace("(", "").replace(")", "").split(", ")
    points = "[[" + "],[".join([p.replace(" ", ", ") for p in points]) + "]]"
    # we need to take from x/y to lat/long
    points = [[x[1], x[0]] for x in eval(points)]

    line = folium.vector_layers.PolyLine(
        points, popup=f"<b>link_id: {row.link_id}</b>", tooltip=f"{row.link_type}", color="blue", weight=10
    ).add_to(network_links)
```
